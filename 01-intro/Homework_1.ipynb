```python
df_asia = df[df['origin']=='Asia']
df_asia
```

```python
df_selected = df_asia[['vehicle_weight', 'model_year']]
```

```python
df_selected
```

```python
df_top7 = df_selected.head(7)
```

```python
df_top7
```

```python
X = df_top7.values
```

```python
X
```

```python
X.shape
```

```python
XTX = X.T @ X
```

```python
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
```

```python
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
y
```

```python
w = XTX_inv @ X.T @ y
```

```python
w
```

```python
result = w.sum()
```

```python
print("Sum of all elements in w:", result.round(5))
```
